```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
%cd /content/drive/My Drive/cs_172b_project/
```

```python
def plot_log_mel_spectrogram(audio_path):
    # Load the audio file
    y, sr = librosa.load(audio_path)

    # Generate a Mel-scaled power spectrogram
    mel_spect = librosa.feature.melspectrogram(y=y, sr=sr)

    # Convert the power spectrogram to dB scale
    log_mel_spect = librosa.power_to_db(mel_spect)

    # Display the spectrogram
    plt.figure(figsize=(8, 2))
    librosa.display.specshow(log_mel_spect, sr=sr, x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Log-Mel spectrogram')
    plt.tight_layout()
    plt.show()
```

```python
# Your directory path where the .wav files are stored
dir_path = './'

# Iterate through your files
for file_name in os.listdir(dir_path):
    if file_name.endswith('.wav'):
        print(f"Processing: {file_name}")
        plot_log_mel_spectrogram(os.path.join(dir_path, file_name))
```

```python
random_clip = 'PSML Nexus Bootcamp - Introduction to Deep Learning - Prof Pierre Baldi, UCI I&CS_clip_48_1493.31_1507.69.wav'
plot_log_mel_spectrogram(os.path.join('./full_pierre_dataset', random_clip))
```
